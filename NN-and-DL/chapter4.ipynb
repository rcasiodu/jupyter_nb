```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class neuro_network(object):
    def __init__(self, sizes):
        self.num_layers = len(sizes)-2
        self.sizes = sizes
        self.weights = [np.random.randn(j, w) for j,w in zip(sizes[1:], sizes[:-1])] 
        self.biases = [np.random.randn(y, 1) for y in sizes[1:]]
    
    def init_wb(self, weights, biases):
        self.weights = weights
        self.biases = biases
    
    def feedforward_output(self, a):
        for weight, bias in zip(self.weights, self.biases):
            z_ = np.dot(weight, a)+bias
            a = self.sigmoid(z_)
        
        return a
    
    def feedforward_layer(self, a, layer):
        for weight, bias in zip(self.weights[:layer], self.biases[:layer]):
            z_ = np.dot(weight, a)+bias
            a = self.sigmoid(z_)
        
        return a    
    
    def net_output(self, a, layer):
        z_ = np.dot(self.weights[layer], a)+self.biases[layer]
        return z_
    
    def sigmoid(self, z_):
        activate_z = 1.0/(1.0+np.exp(-z_))
        return activate_z
```

# Weighted output from hidden layer

```python
dnn = neuro_network([1,2,1])
weights = [np.array([[1000.0],[1000.0]]), np.array([[0.6, 1.2]])]
biases = [np.array([[-400.0],[-600.0]]), np.array([[0.0]])]
dnn.init_wb(weights, biases)

x = np.arange(0, 1.0, 0.01)[:,None]
z_ = []
for j in range(len(x)):
    i = x[j].reshape(1, -1)
    a1 = dnn.feedforward_layer(i, -1)
    z2 = dnn.net_output(a1, -1)
    z_.append(z2[0])

plt.plot(x, z_, color = 'y')
plt.title('Weighted output from hidden layer')
plt.xlabel('x')
plt.ylim([-1, 2])
plt.show()
```
